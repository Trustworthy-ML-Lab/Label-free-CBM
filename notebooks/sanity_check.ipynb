```python
from sentence_transformers import SentenceTransformer, util
import data_utils

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
classes = open("data/nih14_classes.txt").read().splitlines()
concepts = open("data/concept_sets/nih14_biomedclip.txt").read().splitlines()

class_emb = model.encode(classes, normalize=True)
concept_emb = model.encode(concepts, normalize=True)

sim_matrix = util.cos_sim(class_emb, concept_emb)
# inspect max per class
for i, cls in enumerate(classes):
    best = sim_matrix[i].max().item()
    if best > 0.6:  # pick your own threshold
        idx = sim_matrix[i].argmax().item()
        print(f"{cls} overlaps with concept '{concepts[idx]}' (sim={best:.2f})")
```
